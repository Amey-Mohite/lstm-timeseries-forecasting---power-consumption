```python
#import libraries
import pandas as pd
import pytorch_lightning as pl
from pytorch_lightning.callbacks import EarlyStopping
import torch

from pytorch_forecasting.data import NaNLabelEncoder
from pytorch_lightning.callbacks import EarlyStopping, LearningRateMonitor
from pytorch_lightning.loggers import TensorBoardLogger

from pytorch_forecasting import Baseline, TemporalFusionTransformer, TimeSeriesDataSet
from pytorch_forecasting.data import GroupNormalizer
from pytorch_forecasting.metrics import SMAPE, PoissonLoss, QuantileLoss
from pytorch_forecasting.models.temporal_fusion_transformer.tuning import optimize_hyperparameters

import pickle
import os
import warnings
warnings.filterwarnings("ignore")
```

```python
#read dataframe
df = pd.read_csv('/content/drive/MyDrive/Vol(A1487).csv')
df['Date'] = pd.to_datetime(df['Date'])
df['Count'] = [1] * len(df)
df.drop(['Variant_Change', 'Weekly'], axis=1, inplace=True)
df.head()
```

```python
df = df.groupby(['VariantID', 'Date']).sum().reset_index()
df.head()
```

```python
df.shape
```

```python
#convert object to datetime
df["Date"] = pd.to_datetime(df["Date"])
```

```python
#sort dataframe by date
df.sort_values(by='Date', inplace=True)
df.head()
```

```python
df.shape
```

```python
df = df.groupby(['VariantID', pd.Grouper(key='Date', freq='W-MON')])[['CycleTime (s)',	'Downtime (s)',	'Count']].sum().reset_index().sort_values('Date')
df.shape
```

```python
#removing some Variant which has less number of samples
for i in [28,   27,    26,   29,    32,    25,    34,    33,    24,    22,    37,    35     ]:
  df = df[df['VariantID'] != i]
df.shape
```

```python
#missing values
df.isna().sum()
```

```python
df.dropna(axis=0, inplace=True)
```

```python
df.shape
```

```python
df['VariantID'].nunique()
```

```python
#taking only first 5000 samples of each variant
final_df = pd.DataFrame()

for id in list(df['VariantID'].unique()):
  final_df = final_df.append(df[df['VariantID'] == id][:48])
final_df.shape
```

```python
#keeps track of time 
final_df['time_idx'] = list(range(48))*df['VariantID'].nunique()
```

```python
#encoder and prediction timestep length
max_prediction_length = 7
max_encoder_length = 24
training_cutoff = final_df["time_idx"].max() - max_prediction_length
```

```python
#data type conversion
final_df["VariantID"] = final_df["VariantID"].astype(str)
```

```python
final_df.head()
```

```python
# create a timeseries generator to fit the model
training = TimeSeriesDataSet(
    final_df[lambda x: x.time_idx <= training_cutoff],
    time_idx="time_idx",
    target="Count",
    group_ids=["VariantID"],
    min_encoder_length=max_encoder_length // 2,  # keep encoder length long (as it is in the validation set)
    max_encoder_length=max_encoder_length,
    min_prediction_length=1,
    max_prediction_length=max_prediction_length,
    static_categoricals=["VariantID"],
    #static_reals=["avg_population_2017", "avg_yearly_household_income_2017"], 
    time_varying_known_categoricals=[],
    #variable_groups={"special_days": special_days},  # group of categorical variables can be treated as one variable
    time_varying_known_reals=["time_idx", 'CycleTime (s)',	'Downtime (s)'],
    
    time_varying_unknown_categoricals=[],
    time_varying_unknown_reals=[
        "Count"
    ],
    target_normalizer=GroupNormalizer(
        groups=["VariantID"], transformation="softplus"
    ),  # use softplus and normalize by group
    add_relative_time_idx=True,
    add_target_scales=True,
    add_encoder_length=True,
)
```

```python
# create validation set (predict=True) which means to predict the last max_prediction_length points in time
# for each series
validation = TimeSeriesDataSet.from_dataset(training, final_df, predict=True, stop_randomization=True)

# create dataloaders for model
batch_size = 128  # set this between 32 to 128
train_dataloader = training.to_dataloader(train=True, batch_size=batch_size, num_workers=0)
val_dataloader = validation.to_dataloader(train=False, batch_size=batch_size * 10, num_workers=0)
```

```python
# configure network and trainer
early_stop_callback = EarlyStopping(monitor="val_loss", min_delta=1e-4, patience=10, verbose=False, mode="min")
lr_logger = LearningRateMonitor()  # log the learning rate
logger = TensorBoardLogger("lightning_logs")  # logging results to a tensorboard

trainer = pl.Trainer(
    max_epochs=30,
    gpus=0,
    weights_summary="top",
    gradient_clip_val=0.1,
    limit_train_batches=30,  # coment in for training, running valiation every 30 batches
    # fast_dev_run=True,  # comment in to check that networkor dataset has no serious bugs
    callbacks=[lr_logger, early_stop_callback],
    logger=logger,
)


tft = TemporalFusionTransformer.from_dataset(
    training,
    learning_rate=0.03,
    hidden_size=16,
    attention_head_size=1,
    dropout=0.1,
    hidden_continuous_size=8,
    output_size=7,  # 7 quantiles by default
    loss=QuantileLoss(),
    log_interval=10,  # uncomment for learning rate finder and otherwise, e.g. to 10 for logging every 10 batches
    reduce_on_plateau_patience=4,
)
print(f"Number of parameters in network: {tft.size()/1e3:.1f}k")
```

```python
trainer.fit(
    tft,
    train_dataloader=train_dataloader,
    val_dataloaders=val_dataloader,
)
```

```python
# parameter tuning
study = optimize_hyperparameters(
    train_dataloader,
    val_dataloader,
    model_path="optuna_test",
    n_trials=2,
    max_epochs=2,
    gradient_clip_val_range=(0.01, 1.0),
    hidden_size_range=(8, 128),
    hidden_continuous_size_range=(8, 128),
    attention_head_size_range=(1, 4),
    learning_rate_range=(0.001, 0.1),
    dropout_range=(0.1, 0.3),
    trainer_kwargs=dict(limit_train_batches=30),
    reduce_on_plateau_patience=4,
    use_learning_rate_finder=False,  # use Optuna to find ideal learning rate or use in-built learning rate finder
)

# save study results - also we can resume tuning at a later point in time
with open("/content/sample_data/test_study.pkl", "wb") as fout:
    pickle.dump(study, fout)

# show best hyperparameters
print(study.best_trial.params)
```

```python
# load the best model according to the validation loss
# (given that we use early stopping, this is not necessarily the last epoch)
best_model_path = trainer.checkpoint_callback.best_model_path
best_tft = TemporalFusionTransformer.load_from_checkpoint(best_model_path)
```

```python
# calcualte mean absolute error on validation set
actuals = torch.cat([y[0] for x, y in iter(val_dataloader)])
predictions = best_tft.predict(val_dataloader)
(actuals - predictions).abs().mean()
```

```python
# raw predictions are a dictionary from which all kind of information including quantiles can be extracted
raw_predictions, x = best_tft.predict(val_dataloader, mode="raw", return_x=True)
```

```python
for idx in range(7):  # plot 10 examples
    best_tft.plot_prediction(x, raw_predictions, idx=idx, add_loss_to_title=True);
```

```python
predictions = predictions.numpy()
predictions = predictions.astype(int)
predictions = predictions.T

result_df = pd.DataFrame(predictions, columns=['variantID '+str(i) for i in final_df['VariantID'].unique()])
result_df
```
