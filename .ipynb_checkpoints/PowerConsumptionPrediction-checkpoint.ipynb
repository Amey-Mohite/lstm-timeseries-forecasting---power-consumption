```python
# Importing packages
import pandas as pd
import math
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, LSTM, Dropout
from sklearn.preprocessing import MinMaxScaler
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error
```

```python
# Importing Data
df_people = pd.read_csv("../input/temperatureprediction/full occ csv.csv")
df_weather = pd.read_csv("../input/temperatureprediction/weather-1.csv")
df_electricity = pd.read_excel("../input/temperatureprediction/consumption.xlsx")
```

```python
# Removing unwanted columns
df_electricity = df_electricity.drop(["H","OPTIMA Half Hourly DATA","Unnamed: 2","CHANNEL TYPE","UNITS"],axis = 1)

# Slicing all data into same shape
df_weather = df_weather.iloc[:1535,:]
df_people = df_people.iloc[:1535,:]
```

```python
# Viewing Dataframe
df_people.head()
```

```python
# Viewing Dataframe
df_electricity
```

```python
# Viewing Dataframe
df_weather_copy = df_weather.copy()
df_weather.head()
```

```python
df_people[["Date"]] = df_people[["Date"]].apply(pd.to_datetime)
df_electricity["READING DATE"] = df_electricity["READING DATE"].apply(pd.to_datetime)
df_weather["READING DATE"] = df_weather["READING DATE"].apply(pd.to_datetime)
```

```python
# Making sure both reading date columns are same
df_weather["READING DATE"] = df_electricity["READING DATE"]
df_people["Date"] = df_weather["READING DATE"]
# Filling NAN values with 0
df_people = df_people.fillna(0)
```

```python
# Dataframe slicing
df_weather_updated = pd.DataFrame()
df_weather_updated["READING DATE"] = df_weather["READING DATE"]
for j,i in enumerate(range(2,50,2)):
  time = ("Weather_"+str(j)+":00")
  df_weather_updated[time] = df_weather.iloc[:,i]

df_electricity_updated = pd.DataFrame()
df_electricity_updated["READING DATE"] = df_electricity["READING DATE"]
for j,i in enumerate(range(2,50,2)):
  time = ("Electricity_"+str(j)+":00")
  df_electricity_updated[time] = df_electricity.iloc[:,i]

df_people_updated = pd.DataFrame()
df_people_updated["READING DATE"] = df_people["Date"]
for j,i in enumerate(range(1,25,1)):
  time = ("People_"+str(j)+":00")
  df_people_updated[time] = df_people.iloc[:,i]
```

```python
# Updated Dataframe
df_people_updated.head()
```

```python
df_weather_updated.head()
```

```python
print("Weather Dataset Size",df_weather_updated.shape)
print("Electricity Dataset Size",df_electricity_updated.shape)
print("people Dataset Size",df_people_updated.shape)
```

```python
# Dropping Data Column
df_weather = df_weather_updated.drop(["READING DATE"],axis =1)
df_electricity = df_electricity_updated.drop(["READING DATE"],axis =1)
df_electricity = df_electricity.fillna(method='ffill')
df_people = df_people_updated.drop(["READING DATE"],axis =1)
```

```python
# Replacing outlier values
for i,j in enumerate(df_people["People_13:00"]):
  if j == 'UID card access problems':
    df_people["People_13:00"][i] = np.nan

for k in df_people.columns:
  for i,j in enumerate(df_people[k]):
    if j == 'Sentry Juno installation - Sentry offline, no data 9th -22nd July' or j == 'Sentry Juno live but no data yet' or j == "Sentry Juno put into offline mode" or j == "#" or j == " ":
      df_people[k][i] = np.nan

df_people = df_people.fillna(method='ffill')
```

```python
# Merging all dataframes
df_final = pd.DataFrame()
j = 0
k = 0
l = 0
m = 0
for i in range(72):
  if i in range(0,72,3):
    df_final[i] = df_weather.iloc[:,k]
    k = k+1
  if i in range(1,72,3):
    df_final[i] = df_electricity.iloc[:,l]
    l = l+1
  if i  in range(2,72,3):
    df_final[i] =df_people.iloc[:,m]
    m = m+1
```

```python
# Slicing X, Y
X = df_electricity.values.ravel().reshape(-1,1)
Y = df_electricity.values.ravel()
```

```python
# Slicing train data i.e. 80% of total data
training_data_len = math.ceil(len(X) * .96254)

# Applying scaling
scaler_close = MinMaxScaler(feature_range=(0, 1))
scaled_data_close = scaler_close.fit_transform(X)

# Slicing
X_train_scaled = scaled_data_close[0:training_data_len, :]
Y_train_len = Y[0:training_data_len]

# Creating X_train, Y_train using for loop with 60 window size
x_train = []
y_train = []
for i in range(60, len(X_train_scaled)):
    x_train.append(X_train_scaled [i-60:i, 0])
    y_train.append(Y_train_len[i])
    if i <= 61:
        print(x_train)
        print(y_train)
        
# COnverting them to array and reshape
x_train, y_train = np.array(x_train), np.array(y_train)       
x_train = np.reshape(x_train, (x_train.shape[0], x_train.shape[1], 1))
```

```python
# LSTM Model
model = Sequential()
model.add(LSTM(128, return_sequences=True ,input_shape=(x_train.shape[1], 1)))
model.add(Dropout(0.2))
model.add(LSTM(88, return_sequences=True))
model.add(LSTM(48, return_sequences=False))
model.add(Dense(128, activation='relu'))
model.add(Dropout(0.2))
model.add(Dense(64, activation='relu'))
model.add(Dense(16, activation='relu'))
model.add(Dense(1, activation='linear'))

optimizer = tf.keras.optimizers.Adam(learning_rate=0.0001)

# Compiling the model
model.compile(loss='mean_squared_error', optimizer=optimizer)
```

```python
# Fiiting data into model
model.fit(x_train, y_train, batch_size=512, epochs=200)
```

```python
# Applying same train functions for test data
test_data = scaled_data_close[training_data_len:, :]
X_test = []
Y_test = []

# Accessin X_test,Y_test
for i in range(60,len(test_data)):
    X_test.append(test_data[i-60:i, 0])
    Y_test.append(test_data[i])

# Converting them to array
X_test, Y_test = np.array(X_test), np.array(Y_test)
X_test = np.reshape(X_test, (X_test.shape[0], X_test.shape[1], 1))
```

```python
# Model Predictions
preds = model.predict(X_test)
```

```python
# Finding RMSE Value
mse = mean_squared_error(Y_test, preds)
```

```python
# Saving predicted results inot Dataframe
train = pd.DataFrame(scaled_data_close[:training_data_len],columns = ["Close"])
valid = pd.DataFrame(scaled_data_close[training_data_len+60:],columns = ["Close"])
preds = [round(i,2) for i in list(preds.ravel())]
valid['Predictions'] = preds
```

```python
# Converting the data into structured way
df_pred = pd.DataFrame()
for i,j in enumerate(range(48)):
    df_pred[i] = list(preds[60:60+30])
    i = i+30
    
# Create Date Column for predictions
date_list = []
for i,j in enumerate(range(14,44)):
    if j > 30:
        j = j-30
        date_list.append(str(j)+"-12-2021")
    else:
        date_list.append(str(j)+"-11-2021")

# Inserting Date column at 0th index
df_pred.insert(loc=0, column='Date', value=date_list)
```
